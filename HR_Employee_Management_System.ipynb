```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("hr_employee.db")

# Enable Foreign Keys
conn.execute("PRAGMA foreign_keys = ON;")

print("Database Connected Successfully")
```

```python
def execute_query(query, params=()):
    cursor = conn.cursor()
    cursor.execute(query, params)
    conn.commit()

def read_query(query):
    return pd.read_sql_query(query, conn)
```

```python
create_departments = """
CREATE TABLE departments(
    department_id INTEGER PRIMARY KEY,
    department_name TEXT NOT NULL UNIQUE
);
"""

execute_query(create_departments)
```

```python
create_job_roles = """
CREATE TABLE job_roles(
    role_id INTEGER PRIMARY KEY,
    role_title TEXT NOT NULL UNIQUE,
    min_salary REAL NOT NULL CHECK(min_salary>=0),
    max_salary REAL NOT NULL CHECK(max_salary>=min_salary)
);
"""

execute_query(create_job_roles)
```

```python
create_employees = """
CREATE TABLE employees(
    employee_id INTEGER PRIMARY KEY,
    employee_name TEXT NOT NULL,
    email TEXT NOT NULL UNIQUE,
    city TEXT,
    joining_date DATE NOT NULL,
    department_id INTEGER,
    role_id INTEGER,
    status TEXT DEFAULT 'Active'
        CHECK(status IN ('Active','Probation','Resigned')),
    FOREIGN KEY(department_id)
        REFERENCES departments(department_id),
    FOREIGN KEY(role_id)
        REFERENCES job_roles(role_id)
);
"""

execute_query(create_employees)
```

```python
create_attendance = """
CREATE TABLE attendance(
    attendance_id INTEGER PRIMARY KEY,
    employee_id INTEGER,
    attendance_date DATE NOT NULL,
    attendance_status TEXT
        CHECK(attendance_status IN ('Present','Absent','Leave','Half Day')),
    FOREIGN KEY(employee_id)
        REFERENCES employees(employee_id),
    UNIQUE(employee_id,attendance_date)
);
"""

execute_query(create_attendance)
```

```python
create_payroll = """
CREATE TABLE payroll(
    payroll_id INTEGER PRIMARY KEY,
    employee_id INTEGER,
    salary_month TEXT NOT NULL,
    basic_salary REAL NOT NULL CHECK(basic_salary>=0),
    bonus REAL DEFAULT 0 CHECK(bonus>=0),
    deductions REAL DEFAULT 0 CHECK(deductions>=0),
    payment_status TEXT
        CHECK(payment_status IN ('Paid','Pending','Hold')),
    FOREIGN KEY(employee_id)
        REFERENCES employees(employee_id),
    UNIQUE(employee_id,salary_month)
);
"""

execute_query(create_payroll)
```

```python
from IPython.display import HTML, display

tables = ["payroll" ,"employees", "attendance", "job_roles", "departments" ]

html_content = "<table style='width:100%'>"
for i in range(0, len(tables), 4): # Change step to 4 for 4 columns
    html_content += "<tr>"
    for j in range(4): # Loop for 4 columns
        if i + j < len(tables):
            table_name = tables[i + j]
            df = read_query(f"PRAGMA table_info({table_name});")
            html_content += f"<td style='width:25%; vertical-align:top; padding:10px;'>" # Adjust width for 4 columns
            html_content += f"<h3>===== {table_name.upper()} TABLE SCHEMA =====</h3>"
            html_content += df.to_html()
            html_content += "</td>"
        else:
            html_content += "<td style='width:25%;'></td>" # Empty cell to maintain layout
    html_content += "</tr>"
html_content += "</table>"

display(HTML(html_content))
```
