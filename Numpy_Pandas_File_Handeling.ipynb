```python
# Task 7: Text File Handling

marks_list = [85, 92, 76, 88, 69, 91, 55, 78, 39, 95]

total_students = len(marks_list)
average_marks = sum(marks_list) / total_students
highest_marks = max(marks_list)
lowest_marks = min(marks_list)
passed_students = len([m for m in marks_list if m >= 40])
failed_students = len([m for m in marks_list if m < 40])

with open("student_summary.txt", "w") as file:
    file.write("Student Performance Summary\n")
    file.write("----------------------------\n")
    file.write(f"Total Students: {total_students}\n")
    file.write(f"Average Marks: {average_marks:.2f}\n")
    file.write(f"Highest Marks: {highest_marks}\n")
    file.write(f"Lowest Marks: {lowest_marks}\n")
    file.write(f"Passed Students: {passed_students}\n")
    file.write(f"Failed Students: {failed_students}\n")

print("student_summary.txt created successfully.")

print("\nReading File\n")

with open("student_summary.txt","r") as file:
    print(file.read())
```

```python
import csv

students = [
[1,"Rahul",22,"Patna","Python",85,92],
[2,"Priya",24,"Delhi","Data Science",92,88],
[3,"Aman",21,"Kolkata","Python",76,81],
[4,"Neha",23,"Mumbai","Analytics",88,95],
[5,"Rohan",25,"Patna","Python",69,72],
[6,"Simran",22,"Delhi","Data Science",91,89],
[7,"Vikash",24,"Patna","Analytics",55,68],
[8,"Pooja",21,"Kolkata","Python",78,84],
[9,"Arjun",23,"Mumbai","Data Science",39,60],
[10,"Sneha",22,"Delhi","Python",95,97]
]

header = [
"Student_ID",
"Name",
"Age",
"City",
"Course",
"Marks",
"Attendance_Percentage"
]

with open("students.csv","w",newline="") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(students)

print("students.csv created successfully.")

print("\nStudent Records\n")

with open("students.csv","r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
import pandas as pd

df = pd.read_csv("students.csv")

print("First Five Rows")
display(df.head())

print("Last Five Rows")
display(df.tail())

print("Shape")
print(df.shape)

print("\nColumns")
print(df.columns)

print("\nInformation")
df.info()

print("\nStatistical Summary")
display(df.describe())
```

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x>40 else "Fail")
```

```python
def grade(marks):

    if marks>=90:
        return "A+"

    elif marks>=75:
        return "A"

    elif marks>=60:
        return "B"

    elif marks>=40:
        return "C"

    else:
        return "Fail"

df["Grade"]=df["Marks"].apply(grade)
```

```python
df
```

```python
df["Attendance_Status"] = df["Attendance_Percentage"].apply(
    lambda x:"Eligible" if x>=75 else "Low Attendance"
)

display(df)
```

```python
display(df.groupby("Course")["Marks"].min())
```

```python
summary = pd.DataFrame({

"Metric":[
"Total Students",
"Average Marks",
"Highest Marks",
"Lowest Marks",
"Passed Students",
"Failed Students",
"Students with Low Attendance",
"Students Scoring Above 80"
],

"Value":[

len(df),

round(df["Marks"].mean(),2),

df["Marks"].max(),

df["Marks"].min(),

len(df[df["Result"]=="Pass"]),

len(df[df["Result"]=="Fail"]),

len(df[df["Attendance_Status"]=="Low Attendance"]),

len(df[df["Marks"]>80])

]

})

display(summary)
```
