# Retail Sales Data Analysis

### 1. Import Libraries

```python
import sqlite3
import pandas as pd
```

### 2. Connect to SQLite Database

```python
conn = sqlite3.connect("retail_sales.db")

cursor = conn.cursor()

print("Database Created Successfully")
```

### 3. Create Customers Table

```python
cursor.execute("""
CREATE TABLE customers(
    customer_id TEXT PRIMARY KEY,
    customer_name TEXT,
    city TEXT,
    age INTEGER,
    gender TEXT
)
""")
```

### 4. Create Products Table

```python
cursor.execute("""
CREATE TABLE products(
    product_id TEXT PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    price REAL
)
""")
```

### 5. Create Orders Table

```python
cursor.execute("""
CREATE TABLE orders(
    order_id TEXT PRIMARY KEY,
    customer_id TEXT,
    product_id TEXT,
    order_date DATE,
    quantity INTEGER,
    sales_channel TEXT,

    FOREIGN KEY(customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY(product_id) REFERENCES products(product_id)
)
""")
```

### 6. Commit Table Creations

```python
conn.commit()

print("Tables Created Successfully")
```

### 11. Retrieve Joined Sales Data (SQL)

```python
query = """
SELECT
    o.order_id,
    c.customer_name,
    c.city,
    p.product_name,
    p.category,
    o.quantity,
    p.price,
    (o.quantity * p.price) AS total_amount,
    o.sales_channel,
    o.order_date
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
JOIN products p
ON o.product_id = p.product_id
ORDER BY c.customer_name;
"""

pd.read_sql(query, conn)
```

### 12. Calculate Total Revenue (SQL)

```python
query = """
select
sum(o.quantity * p.price) as total_revenue
from orders o
join products p
on o.product_id = p.product_id;
"""

pd.read_sql(query, conn)
```

### 13. Analyze Revenue by City (SQL)

```python
query = """
SELECT
    c.city,
    SUM(o.quantity * p.price) AS total_revenue
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
JOIN products p
ON o.product_id = p.product_id
GROUP BY c.city
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

### 14. Find Best-Selling Product (SQL)

```python
query = """
SELECT
    p.product_name,
    SUM(o.quantity) AS total_quantity_sold
FROM orders o
JOIN products p
ON o.product_id = p.product_id
GROUP BY p.product_name
ORDER BY total_quantity_sold DESC
LIMIT 1;
"""

pd.read_sql(query, conn)
```

### 15. Analyze Category-wise Revenue (SQL)

```python
query = """
SELECT
    p.category,
    SUM(o.quantity * p.price) AS total_revenue
FROM orders o
JOIN products p
ON o.product_id = p.product_id
GROUP BY p.category
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

### 16. Analyze Revenue by Sales Channel (SQL)

```python
query = """
SELECT
    o.sales_channel,
    SUM(o.quantity * p.price) AS total_revenue
FROM orders o
JOIN products p
ON o.product_id = p.product_id
GROUP BY o.sales_channel
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

### 17. Analyze Monthly Revenue Trend (SQL)

```python
query = """
SELECT
    strftime('%Y-%m', o.order_date) AS month,
    SUM(o.quantity * p.price) AS total_revenue
FROM orders o
JOIN products p
ON o.product_id = p.product_id
GROUP BY month
ORDER BY month;
"""

pd.read_sql(query, conn)
```

### 18. Identify High-Value Customers (SQL)

```python
query = """
SELECT
    c.customer_name,
    c.city,
    SUM(o.quantity * p.price) AS total_purchase_value
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
JOIN products p
ON o.product_id = p.product_id
GROUP BY c.customer_name, c.city
HAVING SUM(o.quantity * p.price) > 50000
ORDER BY total_purchase_value DESC;
"""

pd.read_sql(query, conn)
```

### 19. Load Data into Pandas DataFrames

```python
import pandas as pd

customers_df = pd.read_sql("SELECT * FROM customers", conn)
products_df = pd.read_sql("SELECT * FROM products", conn)
orders_df = pd.read_sql("SELECT * FROM orders", conn)

print("Customers DataFrame")
display(customers_df)

print("Products DataFrame")
display(products_df)

print("Orders DataFrame")
display(orders_df)
```

### 20. Merge DataFrames to Create Sales DataFrame

```python
sales_df = pd.merge(
    orders_df,
    customers_df,
    on="customer_id",
    how="inner"
)

sales_df = pd.merge(
    sales_df,
    products_df,
    on="product_id",
    how="inner"
)

print("Merged Sales DataFrame")
display(sales_df)
```

### 21. Calculate Total Amount for Each Order

```python
sales_df["total_amount"] = sales_df["quantity"] * sales_df["price"]

print("Sales DataFrame with Total Amount")
display(sales_df)
```

### 22. Calculate Total Revenue (Pandas)

```python
total_revenue = sales_df["total_amount"].sum()

print("Total Revenue")
print(total_revenue)
```

### 23. Analyze Revenue by City (Pandas)

```python
city_revenue = (
    sales_df
    .groupby("city")["total_amount"]
    .sum()
    .reset_index()
    .sort_values(by="total_amount", ascending=False)
)

print("Revenue by City")
display(city_revenue)
```

### 24. Analyze Product-wise Quantity Sold (Pandas)

```python
product_sales = (
    sales_df
    .groupby("product_name")["quantity"]
    .sum()
    .reset_index()
    .sort_values(by="quantity", ascending=False)
)

print("Product-wise Quantity Sold")
display(product_sales)
```

### 25. Analyze Category-wise Revenue (Pandas)

```python
category_revenue = (
    sales_df
    .groupby("category")["total_amount"]
    .sum()
    .reset_index()
    .sort_values(by="total_amount", ascending=False)
)

print("Category-wise Revenue")
display(category_revenue)
```

### 26. Analyze Revenue by Sales Channel (Pandas)

```python
channel_revenue = (
    sales_df
    .groupby("sales_channel")["total_amount"]
    .sum()
    .reset_index()
    .sort_values(by="total_amount", ascending=False)
)

print("Revenue by Sales Channel")
display(channel_revenue)
```

### 27. Analyze Monthly Revenue Trend (Pandas)

```python
sales_df["order_date"] = pd.to_datetime(sales_df["order_date"])

sales_df["month"] = sales_df["order_date"].dt.strftime("%Y-%m")

monthly_revenue = (
    sales_df
    .groupby("month")["total_amount"]
    .sum()
    .reset_index()
    .sort_values(by="month")
)

print("Monthly Revenue Trend")
display(monthly_revenue)
```

### 28. Identify High-Value Customers (Pandas)

```python
high_value_customers = (
    sales_df
    .groupby(["customer_name", "city"])["total_amount"]
    .sum()
    .reset_index()
)

high_value_customers = high_value_customers[
    high_value_customers["total_amount"] > 50000
]

high_value_customers = high_value_customers.sort_values(
    by="total_amount",
    ascending=False
)

print("High-Value Customers")
display(high_value_customers)
```

### 29. Final Sales DataFrame

```python
print("Final Sales DataFrame")
display(sales_df)
```
